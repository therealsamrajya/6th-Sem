Importing of libraries

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Loading and preprocessing

```python
movie_data = pd.read_csv("/content/movies.csv")
```

```python
movie_data.head()
```

```python
movie_data.shape
```

```python
movie_data.describe()
```

selecting the columns that are used for the recommendation to be done

```python
selected_features=['genres','keywords','tagline','cast','director']
print(selected_features)
```

replace the null values with null string. Null string is a empty string(' '). so fillna means null values or not available with empty string.
(fillna) : feature is a variable that
changes like the  iteration and everytime it takes the value as genres,keywords in a loop.
second statement change the feature value which is numm of particular column with null string.
fillna('n/a')=not avaliable will be filled in empty places

```python
for feature in selected_features:
  movie_data[feature]=movie_data[feature].fillna('')
```

all features with their data are combined and printed as one value

feature will take the value as genre, keyword, tagline etc and one by one all are filled with null string when value is not available or value is missiing

```python
movie_data.head()
```

```python
combined_features = movie_data['genres']+''+movie_data['keywords']+''+movie_data['tagline']+''+movie_data['cast']+''+movie_data['director']
```

concatenation is done

```python
print(combined_features)
```

now converting the data to the numeric data

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

now we can see text data is convertted from text data to numeric data

now try to get similarity score usine cosine similarity funcction

```python
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
print('movies suggested for you: \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=movie_data[movie_data.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1
```

movie recommendation system

```python
movie_name=input("enter your favourite movie name")

list_of_all_titles=movie_data['title'].tolist()

find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)

close_match=find_close_match[0]

index_of_the_movie = movie_data[movie_data.title==close_match]['index'].values[0]

similarity_score=list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)

print('movies suggested for you: \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=movie_data[movie_data.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1

```
